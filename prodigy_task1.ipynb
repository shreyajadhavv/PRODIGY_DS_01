```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Loan - Loan.csv")
df.head()

```

```python
#histogram that shows how loan amount is distributed
plt.figure(figsize=(8,5))
plt.hist(df['disbursed_amount'], bins=20)

plt.title("Distribution of Loan Amount")

plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#barchart shows why people take loans
plt.figure(figsize=(8,5))
df['loan_purpose'].value_counts().plot(kind='bar')
plt.title("Loan Purpose Distribution")
plt.xlabel("Loan purpose")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

df['employment'].value_counts().plot(kind='bar')

plt.title("Employment Distribution")
plt.xlabel("Employment Type")
plt.ylabel("Number of Customers")

plt.show()
```
